````python
#Métodos para operaciones lógicas con matrices

print(f'Greater than 1 AND less than 5:\n{np.logical_and(a> 1, a<5)}\n')
print(f'{(a>1)&(a<5)}\n') #ES equivalente
print(f'Greater than 3 OR less than 4:\n{np.logical_or(a> 3, a<2)}\n')
print(f'Greater than 1 NOT less than 5:\n{np.logical_and(a> 1, a>=5)}\n')
print(f'Greater than 1 XOR less than 5:\n{np.logical_xor(a> 1, a<5)}\n')
print(f'Usando el logical_not:\nLa matriz:\n {a>1},\n y su negación:\n {np.logical_not(a> 1)}\n')
````

````python
#Como ya hemos indicado, se dispone de operadores relacionales entre arrays!!

array_one = np.array([1, 1, 3])
array_two = np.array([1, 2, 2])

print(f'Element-wise GREATER than comparison:\n{np.greater(array_one, array_two)}\n')
#No puede ser de otro modo, lo visto antes es sólo un caso particular de esto. Por ejemplo:
print(f'Element-wise GREATER than comparison:\n{np.greater(array_one, 2)}\n')
print(f'Element-wise GREATER than comparison:\n{array_one>2}\n')
print(f'Element-wise GREATER than comparison shorthand (>):\n{array_one > array_two}\n')
print(f'Element-wise GREATER than comparison (flip):\n{np.greater(array_two, array_one)}\n')
print(f'Element-wise GREATER than comparison shorthand (>) (flip):\n{array_two > array_one}')
````

Se dispone de una expresión condicional generalizada a arrays, del mismo tipo de la que se tiene con expresiones de tipos básicos en los lenguajes de programación. Para ello se usa el método where.

````python
xarr = np.array([1.1, 1.2, 1.3, 1.4])
yarr = np.array([2.1, 2.2, 2.3, 2.4])
cond = np.array([True, False, True, False])
result = np.where(cond, xarr, yarr)
print(result)
````

````python
# Hecho "a pelo" usando listas por comprensión y la expresión condicional de Python
result = [(x if c else y) for x, y, c in zip(xarr, yarr, cond)]
print(result)
````

````python
#Los dos últimos argumentos de where pueden no ser arrays
a=np.array([[True,True],[False,True]])
print(np.where(a,0,1))
````

## Indexado de arrays y slicing

Para seguir esta segunda parte de la sesión 2 de Numpy, usaremos [la Sección 1 del Cap. 4 del libro de referencia](https://wesmckinney.com/book/numpy-basics#ndarray_indexing) y el [tutorial](https://github.com/pabloinsente/intro-sc-python/blob/master/notebooks/intro-numpy-fundamentals.ipynb).

Muchos de los ejemplos que siguen están sacados de la documentación.

Comenzamos tratando el tema de modo general, para luego centrarnos en aspectos que nos pueden resultar útiles en tratamiento de datos.

Cada elemento de un array tiene asociado su índice, una tupla de enteros que indican su posición en cada eje del array.

Como es natural, para indexar en Python usaremos []. Nos va a pernitir acceder, mofificar y borrar elementos de un array, ofreciendo toda la capacidad de los slices y herramientas de más alto nivel.

### Indexado básico

````python
#Sobre arrays unidimensionales todo sigue la misma norma que en Python
#Acceso por índice

array_one = np.arange(1,11)

print(f'Array one: {array_one}')
print(f'Array one dimensions: {array_one.ndim}, shape:{array_one.shape}')
print(f'Select element at position [0]: {array_one[0]}')
print(f'Select element at position [5]: {array_one[5]}')
print(f'Select element at position [-1]: {array_one[-1]}')#Sobre arrays unidimensionales todo igue la misma norma que en Python
````

````python
#Slices
print(f'Elements from position [0] to position [3]: {array_one[0:3]}')
print(f'Elements from position [5] to position [9]: {array_one[5:9]}')
print(f'Elements from position [-9] to position [-5]: {array_one[-9:-5]}')
print(f'Elements from position [-3] to position [-1]: {array_one[-3:-1]}')
print(f'Elements from position [3] to position [-1]: {array_one[3:-1]}')
print(f'\n')
#Se puede añadir el "paso"
print(f'Slice from position [0] to position [6] with stride [1]: {array_one[0:6:1]}')
print(f'Slice from position [0] to position [6] with stride [2]: {array_one[0:6:2]}')
print(f'Slice from position [-6] to position [-1] with stride [3]: {array_one[-6:-1:3]}')
````

Para el indexado sobre arrays multidimensionales hay dos formas:

- Entre corchetes,índices de cada eje separados por comas. Esto es: ``[,]``
- Cada dimensión con su correspondiente pareja de corchetes. Esto es: ``[][]``


````python
#Indexado sencillo
array_two = np.arange(1,10).reshape((3,3))
array_three = np.arange(1,9).reshape((2,2,2))

print(f'Array two dimensions/axes: \n{array_two}\n')
print(f'Array three dimensions/axes: \n{array_three}\n')
print(f'Array two dimensions: {array_two.ndim}, shape:{array_two.shape}')
print(f'Array three dimensions: {array_three.ndim}, shape:{array_three.shape}')
````

````python
#Acceso a elementos usando las dos formas
print(f'Element at position 1 in first axis (rows) and position 1 in second axis (cols): {array_two[1,1]}')
print(f'Element at position -1 in first axis (rows) and position -3 in second axis (cols): {array_two[-1,-3]}')
print(f'\n')
print(f'Element at position 1 in first axis (rows) and position 1 in second axis (cols): {array_two[1][1]}')
print(f'Element at position -1 in first axis (rows) and position -3 in second axis (cols): {array_two[-1][-3]}')
````

````python
#Indexado combinado con slices
#Observar el cambio de dimensión
print(f'Array two as reference: \n{array_two}\n')

print(f'All elements at position 0 from first axis (all elements from first row): \n{array_two[0,:]}\n')
#Es equivalente a:
print(f'All elements at position 0 from first axis (all elements from first row): \n{array_two[0][:]}\n')

print(f'All elements at position 2 from first axis (all elements from third row): \n{array_two[2,:]}\n')
print(f'All elements at position 0 from second axis (all elements from first  column): \n{array_two[:,0]}\n')
print(f'All elements at position 1 from second axis (all elements from second column): \n{array_two[:,1]}\n')
````

````python
#Como se espera, podemos conseguir submatrices (tipo caja)
array_two = np.arange(1,10).reshape((3,3))


print(f'Array two dimensions/axes: \n{array_two}\n')
print(f'Elements at position 0 and 1 from first axis and position 0 from second axis: \n{array_two[0:2,0]}\n')
print(f'Elements at position 0 and 1 from first axis and position 0 and 1 from second axis: \n{array_two[0:2,0:2]}\n')
#Observar el uso mediante acceso "curryficando" (https://es.wikipedia.org/wiki/Currificaci%C3%B3n)
print(f'Elements at position 0 and 1 from first axis and position 0 and 1 from second axis: \n{array_two[0:2][:,0:2]}\n')
#no daría lo mismo
print(f'Error de cálculo:\nElements at position 0 and 1 from first axis and position 0 and 1 from second axis:\n{array_two[0:2][0:2]}\n')

print(f'Elements at position 1 and 2 from first axis and position 1 and 2 from second axis: \n{array_two[1:3,1:3]}\n')

print('Con paso\n')
print(f'All elements at position 1 and 2 from first axis and second axis with stride 2: \n{array_two[0:2,::2]}\n')
````

````python
#submatrices que no son cajas: calcular la submatriz dada por filas 0 y 2 y columnas 0 y 2
print(array_two[[0,2]][:,[0,2]])
````

### Indexado booleano (máscara)

Se puede usar un array de booleanos para poder indexar sobre otro (que tenga la misma configuración).
Los elementos true de la máscara seleccionan los elementos accesibles. Es una herramienta esencial en análisis de datos ya que permite "separar" los datos que cumplen unas determinadas condiciones.

````python
#Combinado con aritmética
array = np.array([1,56,-4,6,8,2])
mask = array > 6

print(f'Array: \n{array}\n')
print(f'Mask or Boolean array with "True" for values strictly grater than 6: \n{mask}\n')
#Lo podemos usar para filtrar en el propio array
print(f'Return an sub-array where "mask" elements are "True": \n{array[mask]}\n')
````

````python
print(f'Boolean array with elements larger than 1 and smaller than 9:\n{[(1 < array) & (array < 9)]}\n')
print(f'Select elements larger than 1 and smaller than 9:\n{array[(1 < array) & (array < 9)]}')
print(f'Select elements equal to 2 OR larger than 9 :\n{array[(2 == array) | (array > 9)]}\n')
print(f'Select even elements (modulo == 0) OR larger than 9 :\n{array[((array % 2) == 0) | (array > 9)]}\n')
````

````python
#Con varias dimensiones...
#Se aplana el resultado
array_two = np.arange(1,17).reshape((4,4))
mask_two =  array_two > 8

print(f'Two-dimensional array: \n{array_two}\n')
print(f'Mask or Boolean arrayw with "True" for values strictly greater than 8: \n{mask_two}\n')
print(f'Return an sub-array where "mask" elements are "True": \n{array_two[mask_two]}\n')
print(f'Shape new sub-array: {array_two[mask_two].shape}')
````

````python
mask_three = np.array([True, False, False, True])
mask_four = np.array([False, True, True, False])

print(f'Two-dimensional array: \n{array_two}\n')
print(f'Select elements at position 1 and 4 along the first axis (first and last rows): \n{array_two[mask_three]}\n')
print(f'Select elements at position 2 and 3 along the second axis (second and third cols): \n{array_two[:,mask_four]}\n')
````
